```python
import pandas as pd
df = pd.read_csv('pinguinos.csv')
df
```

```python
df_espanol = df.rename(columns={
    'species': 'especie',
    'island': 'isla',
    'culmen_length_mm': 'longitud_pico_mm',
    'culmen_depth_mm': 'profundidad_pico_mm',
    'flipper_length_mm': 'longitud_aleta_mm',
    'body_mass_g': 'masa_corporal_g',
    'sex': 'sexo'
})

display(df_espanol)
```

```python
df = df.dropna(subset=['flipper_length_mm'], axis=0)
df
```

```python
import matplotlib.pyplot as plt

df.boxplot(column= "flipper_length_mm", by = "sex")
plt.title("Longitud de la Aleta por Sexo")
plt.xlabel("Sexo")
plt.ylabel("Longitud de la Aleta (mm)")
plt.show()
```

```python
import seaborn as sns
sns.boxplot(df, x= "sex", y = "flipper_length_mm"). set(title = "Longitud de la Aleta vs Sexo")
plt.xlabel("Sexo")
plt.ylabel("Longitud de la Aleta (mm)")
```

```python
# Violin Plot con Seaborn
# https://seaborn.pydata.org/generated/seaborn.violinplot.html
sns.violinplot(df, x='sex', y='flipper_length_mm').set(title='Sexo vs Longitud de la Aleta')
plt.xlabel("Sexo")
plt.ylabel("Longitud de la Aleta (mm)")
```

```python
# grafico de barras
df.groupby('sex')['flipper_length_mm'].mean().plot(kind='bar')
plt.title("Longitud Promedio de la Aleta por Sexo")
plt.xlabel("Sexo")
plt.ylabel("Longitud Promedio de la Aleta (mm)")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lmplot(data=df, x='flipper_length_mm', y='body_mass_g', hue='species', palette='viridis', scatter_kws={'s': 32, 'alpha': 0.8})
plt.title('Longitud de la Aleta vs Masa Corporal por Especie con Líneas de Regresión')
plt.xlabel("Longitud de la Aleta (mm)")
plt.ylabel("Masa Corporal (g)")
plt.show()
```

```python
import statsmodels.formula.api as smf

# Get the unique species
species = df['species'].unique()

# Iterate over each species and fit a linear regression model
for s in species:
    # Filter the dataframe for the current species
    df_species = df[df['species'] == s]

    # Fit the linear regression model
    model = smf.ols('body_mass_g ~ flipper_length_mm', data=df_species).fit()

    # Print the equation
    print(f"Ecuación para {s}:")
    print(f"body_mass_g = {model.params['Intercept']:.2f} + {model.params['flipper_length_mm']:.2f} * flipper_length_mm")
    print("-" * 30)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df.drop(columns=['species', 'island', 'sex']).corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación de Variables Numéricas')
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
correlation_matrix = df.drop(columns=['species', 'island', 'sex']).corr()

# Rename columns and index to Spanish using df_espanol column names
spanish_column_names = {
    'culmen_length_mm': 'longitud_pico_mm',
    'culmen_depth_mm': 'profundidad_pico_mm',
    'flipper_length_mm': 'longitud_aleta_mm',
    'body_mass_g': 'masa_corporal_g'
}
correlation_matrix = correlation_matrix.rename(columns=spanish_column_names, index=spanish_column_names)

# Create a mask for the upper triangle, including the diagonal
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool), k=1)

# Create annotations with asterisks for strong correlations
annot_matrix = correlation_matrix.copy().astype(str)
for i in range(correlation_matrix.shape[0]):
    for j in range(correlation_matrix.shape[1]):
        if not mask[i, j]: # Only annotate the visible part
            value = correlation_matrix.iloc[i, j]
            if abs(value) >= 0.7:
                annot_matrix.iloc[i, j] = f"{value:.2f}*"
            else:
                 annot_matrix.iloc[i, j] = f"{value:.2f}"


plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, mask=mask, annot=annot_matrix, fmt="", cmap='coolwarm') # Use fmt="" because annot_matrix is already formatted
plt.title('Matriz de Correlación con Correlaciones Fuertes Marcadas')
plt.show()
```

## Gráficos de caja y bigotes (box plots)

### Subtask:
Generar un gráfico de caja y bigotes para visualizar la distribución de una variable numérica para diferentes categorías (por ejemplo, la longitud de la aleta por especie) y explicar el código y el gráfico.


**Reasoning**:
Generate a box plot to visualize the distribution of flipper length by species and explain the code and the plot in Spanish.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a box plot using seaborn
# We use seaborn.boxplot() to create the box plot.
# 'species' is set on the x-axis to represent different categories (species).
# 'flipper_length_mm' is set on the y-axis to show the distribution of flipper length for each species.
# The data source is the DataFrame 'df'.
sns.boxplot(data=df, x='species', y='flipper_length_mm')

# Set the title of the plot
# We add a descriptive title to the plot in Spanish.
plt.title("Distribución de Longitud de Aleta por Especie")

# Label the x-axis and y-axis
# We label the x-axis as "Especie" and the y-axis as "Longitud de la Aleta (mm)" for clarity.
plt.xlabel("Especie")
plt.ylabel("Longitud de la Aleta (mm)")

# Display the plot
# Finally, we display the generated plot.
plt.show()

```


**Explicación del Código y el Gráfico de Caja y Bigotes:**

Este gráfico de caja y bigotes (boxplot) visualiza la distribución de la longitud de la aleta (en mm) para cada una de las especies de pingüinos en el conjunto de datos.

**Código:**

1.  `import matplotlib.pyplot as plt` y `import seaborn as sns`: Estas líneas importan las librerías necesarias para crear gráficos. `matplotlib.pyplot` se utiliza para personalizar los gráficos (como añadir títulos y etiquetas), y `seaborn` se utiliza para crear gráficos estadísticos de manera más sencilla.
2.  `sns.boxplot(data=df, x='species', y='flipper_length_mm')`: Esta es la función principal que crea el boxplot.
    *   `data=df`: Especifica que los datos a utilizar provienen del DataFrame `df`.
    *   `x='species'`: Indica que la columna 'species' se utilizará para definir las categorías en el eje X (cada caja representará una especie diferente).
    *   `y='flipper_length_mm'`: Indica que la columna 'flipper_length_mm' se utilizará para mostrar la distribución de los valores en el eje Y.
3.  `plt.title(...)`, `plt.xlabel(...)`, `plt.ylabel(...)`: Estas líneas añaden un título al gráfico y etiquetas a los ejes X e Y para hacerlo más comprensible.
4.  `plt.show()`: Esta línea muestra el gráfico generado.

**Gráfico de Caja y Bigotes:**

Cada caja en el gráfico representa la distribución de la longitud de la aleta para una especie específica de pingüino.

*   **La Caja (Box):** La caja central abarca desde el primer cuartil (Q1) hasta el tercer cuartil (Q3) de los datos. Esto significa que el 50% central de los datos se encuentra dentro de esta caja.
*   **La Línea dentro de la Caja:** La línea horizontal dentro de la caja representa la mediana (el segundo cuartil, Q2) de los datos. Es el punto que divide la distribución en dos mitades iguales.
*   **Los Bigotes (Whiskers):** Las líneas que se extienden desde la caja (los "bigotes") se extienden hasta el valor más bajo y el valor más alto dentro de 1.5 veces el rango intercuartílico (IQR, que es Q3 - Q1). Representan el rango típico de los datos, excluyendo los valores atípicos.
*   **Los Puntos (Outliers):** Los puntos individuales que se encuentran fuera de los bigotes se consideran valores atípicos (outliers). Estos son puntos de datos que están significativamente alejados de la mayoría de los otros datos en esa categoría.

En este gráfico específico, puedes comparar visualmente la distribución de la longitud de la aleta entre las diferentes especies de pingüinos. Por ejemplo, puedes ver la mediana de la longitud de la aleta para cada especie, la dispersión de los datos (representada por el tamaño de la caja y los bigotes), y si hay valores de longitud de aleta inusualmente altos o bajos para alguna especie.

## Diagrama de torta (pie chart)

### Subtask:
Generar un diagrama de torta para mostrar la proporción de diferentes categorías en una variable (por ejemplo, la proporción de cada especie) y explicar el código y el gráfico.


**Reasoning**:
Calculate the counts for each species and create a pie chart to visualize the proportion of each species, along with a title and equal axis to ensure it's a circle.



```python
# Calcular el número de ocurrencias de cada especie
# Utilizamos value_counts() en la columna 'species' para contar cuántas veces aparece cada especie.
species_counts = df['species'].value_counts()

# Crear el diagrama de torta
# Usamos plt.pie() para crear el diagrama de torta.
# 'species_counts' proporciona los tamaños de las cuñas.
# 'labels' proporciona las etiquetas para cada cuña (los nombres de las especies).
# 'autopct' formatea y muestra el porcentaje en cada cuña.
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%')

# Añadir título al diagrama de torta
# Añadimos un título descriptivo al diagrama.
plt.title("Proporción de Especies de Pingüinos")

# Asegurar que el diagrama sea un círculo perfecto
# Esto evita que el diagrama se vea elíptico.
plt.axis('equal')

# Mostrar el diagrama
# Finalmente, mostramos el diagrama de torta generado.
plt.show()
```

Print the detailed explanation of the code and the pie chart in Spanish as requested in the instructions.



```python
# Detailed explanation of the code and the pie chart in Spanish
print("""
**Explicación del Código y el Diagrama de Torta:**

Este diagrama de torta visualiza la proporción de cada especie de pingüino en el conjunto de datos. Cada "rebanada" del pastel representa una especie diferente, y el tamaño de la rebanada es proporcional al número de pingüinos de esa especie en comparación con el total.

**Código:**

1.  `species_counts = df['species'].value_counts()`: Esta línea calcula cuántas veces aparece cada valor único en la columna 'species' del DataFrame `df`. El resultado es una Serie donde el índice son los nombres de las especies y los valores son sus recuentos.
2.  `plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%')`: Esta es la función principal que crea el diagrama de torta.
    *   `species_counts`: Proporciona los tamaños de cada "rebanada" del pastel.
    *   `labels=species_counts.index`: Asigna los nombres de las especies (el índice de `species_counts`) como etiquetas para cada rebanada.
    *   `autopct='%1.1f%%'`: Formatea y muestra el porcentaje que cada rebanada representa del total, con un decimal.
3.  `plt.title("Proporción de Especies de Pingüinos")`: Añade un título descriptivo al diagrama.
4.  `plt.axis('equal')`: Asegura que el diagrama de torta se muestre como un círculo perfecto y no distorsionado.
5.  `plt.show()`: Muestra el diagrama de torta generado.

**Diagrama de Torta:**

*   **Las Rebanadas (Wedges):** Cada sección de color en el pastel representa una de las especies de pingüinos (Adelie, Gentoo, Chinstrap).
*   **Las Etiquetas (Labels):** Junto a cada rebanada, verás el nombre de la especie a la que corresponde.
*   **Los Porcentajes:** Dentro o junto a cada rebanada, se muestra el porcentaje que esa especie representa del total de pingüinos en el conjunto de datos.

Observando el diagrama, puedes ver rápidamente qué especie es la más abundante, cuál es la menos común y la proporción relativa de cada especie en el conjunto de datos.
""")
```

## Histogramas con campana de gauss

### Subtask:
Generar un histograma para una variable numérica y superponer una curva de distribución normal (campana de Gauss) para mostrar cómo se distribuyen los datos y explicar el código y el gráfico.


**Reasoning**:
Generate a histogram with a superimposed KDE curve for a numerical variable, add titles and labels in Spanish, display the plot, and then print an explanation of the code and the plot.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select a numerical variable
variable_name = 'flipper_length_mm'
spanish_variable_name = 'Longitud de la Aleta' # Spanish name for the variable

# Generate the histogram with KDE
# We use seaborn.histplot() to create the histogram.
# 'data=df' specifies the DataFrame.
# 'x=variable_name' specifies the numerical variable for the histogram.
# 'kde=True' superimposes the Kernel Density Estimate curve (Gaussian-like curve).
sns.histplot(data=df, x=variable_name, kde=True)

# Add a descriptive title in Spanish
plt.title(f"Distribución de {spanish_variable_name} con Curva KDE")

# Label the axes in Spanish
plt.xlabel(f"{spanish_variable_name} (mm)")
plt.ylabel("Frecuencia")

# Display the plot
plt.show()

# Print explanation of the code and the plot in Spanish
print(f"""
**Explicación del Código y el Histograma con Curva KDE:**

Este gráfico es un histograma que muestra la distribución de la variable "{spanish_variable_name}" (Longitud de la Aleta en mm) en el conjunto de datos. Además, se superpone una curva de Estimación de Densidad del Kernel (KDE), que es una forma suavizada de la distribución y a menudo se asemeja a una campana de Gauss si los datos tienen una distribución aproximadamente normal.

**Código:**

1.  `import matplotlib.pyplot as plt` y `import seaborn as sns`: Importan las bibliotecas necesarias para graficar.
2.  `variable_name = 'flipper_length_mm'`: Define la variable numérica a analizar.
3.  `spanish_variable_name = 'Longitud de la Aleta'`: Define el nombre en español para usar en los títulos y etiquetas.
4.  `sns.histplot(data=df, x=variable_name, kde=True)`: Crea el histograma.
    *   `data=df`: Usa el DataFrame `df`.
    *   `x=variable_name`: Especifica la columna para el histograma.
    *   `kde=True`: Activa la superposición de la curva KDE.
5.  `plt.title(...)`, `plt.xlabel(...)`, `plt.ylabel(...)`: Añaden el título y las etiquetas a los ejes en español.
6.  `plt.show()`: Muestra el gráfico.

**Histograma con Curva KDE:**

*   **Las Barras (Histograma):** Cada barra vertical representa un rango de valores de la longitud de la aleta. La altura de la barra indica cuántos pingüinos tienen una longitud de aleta dentro de ese rango (la frecuencia). El histograma te permite ver la forma general de la distribución de los datos.
*   **La Curva KDE (Línea Suavizada):** La línea azul que se superpone al histograma es la curva de Estimación de Densidad del Kernel. Es una representación suavizada de la distribución de los datos. Si los datos se distribuyen de manera similar a una distribución normal (campana de Gauss), esta curva tendrá una forma de campana simétrica. La curva KDE ayuda a visualizar la forma de la distribución de los datos de manera más clara que solo las barras del histograma, especialmente con un tamaño de muestra limitado.

Al observar este gráfico, puedes evaluar si la longitud de la aleta sigue una distribución aproximadamente normal. Una campana simétrica indicaría normalidad, mientras que formas sesgadas o con múltiples picos sugerirían distribuciones diferentes.
""")
```
